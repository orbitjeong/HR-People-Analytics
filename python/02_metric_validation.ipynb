```python
import pandas as pd
```

```python
df = pd.read_csv("people_fact_clean.csv")
```

```python
df["hire_date"] = pd.to_datetime(df["hire_date"], errors="coerce")
```

```python
df["term_date"] = pd.to_datetime(df["term_date"], errors="coerce")
```

```python
df.head()
```

```python
def headcount_as_of(df, as_of_date):
    as_of_date = pd.to_datetime(as_of_date)
    return df[
        (df["hire_date"] <= as_of_date) &
        ((df["term_date"].isna()) | (df["term_date"] > as_of_date))
    ]["employee_id"].nunique()
```
